#### Implement the function `fit_1d_linear_regression` that takes a matrix $x$ and a vector $y$, and returns a numpy array , where the first element $\beta_0$ is the intercept of the line, and remainings are the coefficients of a features, like $\beta_1$, $\beta_2$, etc.

```python
def fit_1d_linear_regression(data_vector, response_vector):
    data_vector['intercept'] = 1
    data_vector1 = data_vector.to_numpy()
    response_vector1 = response_vector.to_numpy()
    B = np.dot(np.linalg.inv(np.dot(np.transpose(data_vector1),data_vector1)),
               np.dot(np.transpose(data_vector1),response_vector))
    return B
```

#### Implement the following functions: mean_squared_error, mean_absolute_error, r2_score.
#### Each of them should have 2 vector arguments and return the appropriate score.

```python
## for simplicity not using manual code above
def mean_squared_error1(data_vector, response_vector):
    regr = linear_model.LinearRegression()
    regr.fit(data_vector, response_vector)
    y_pred = regr.predict(data_vector)## ess estimating on train set, not on test.
    mse = ((y_pred - response_vector) ** 2).mean()
    mabe = abs(y_pred - response_vector).mean()
    r_2 = 1 - mse / response_vector.var()
    return mse,mabe,r_2
    
mean_squared_error1(xTrainfinal, yTrain)   

```
